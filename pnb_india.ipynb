```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm
from pprint import pprint
import pickle
```

```python
url='https://www.pnbindia.in/EAuction.aspx'
driver = webdriver.Chrome()
driver.get(url)
soup = BeautifulSoup(driver.page_source,'lxml')
obj_dict ={}
for i in range(0,40):
    l1='ContentPlaceHolder1_rptGrid_lblTenderName_'+str(i)
    l2='ContentPlaceHolder1_rptGrid_Label1_'+str(i)
    l3='ContentPlaceHolder1_rptGrid_Label2_'+str(i)
    l4='ContentPlaceHolder1_rptGrid_lbtnTenderTitle_'+str(i)
    office=driver.find_element_by_id(l1).text
    area=driver.find_element_by_id(l2).text
    detail=driver.find_element_by_id(l3).text.replace('Uploaded On','')
    date=driver.find_element_by_id(l4).text
    url=driver.find_element_by_id(l4).get_attribute('href')
    obj_dict[date]={
        'Area' : area,
        'Uploaded On' : detail,
        'E-Auction Date/Details' :date,
        'Office': office,
        'url': url
    }
    df=pd.DataFrame(obj_dict).T
print(df)
# l= driver.find_elements_by_tag_name('a')
# for p in l:
#     print(p.text)
driver.close()
```
